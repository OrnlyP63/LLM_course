```python
import urllib.request
```

```python
url = "https://gutenberg.org/cache/epub/35/pg35.txt"
filename = "./data/the_time_machine.txt"

with urllib.request.urlopen(url) as response:
        # Read and decode the content from bytes to a string
        text_content = response.read().decode('utf-8')

# Open the file in write mode ('w') with UTF-8 encoding
with open(filename, "w", encoding="utf-8") as file:
    file.write(text_content)
```
